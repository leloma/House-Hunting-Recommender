```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from lxml import html
```

#For Sale


```python
# Function to get data from all pages inside pagination
def getnextpage(soup):
    page = soup.find('div', {'class': 'pagination-wrap'})

    if not page:
        return None  # If there's no pagination, end the search

    next_page_link = page.find('li', {'class': 'page-item', 'class': 'active'}).find_next('li', {'class': 'page-item'})

    if not next_page_link or 'disabled' in next_page_link.get('class', []):
        return None  # If there's no next page or it's disabled, end the search

    return next_page_link.find('a')['href']
```

```python
for_sale_url = 'https://www.realethio.com/status/for-sale/'
salepage = requests.get(for_sale_url)
soup = BeautifulSoup(salepage.text, 'html.parser')

# Initialize an empty dataframe
# Define the order of columns
columns_order = ['Property URL', 'Title', 'Type', 'Price', 'Address', 'Description',
                  'Property Size', 'Land Area', 'Bedrooms', 'Bathrooms', 'Garages',
                  'Property Type', 'Property Status', 'Floor', 'Furnished']

# Initialize an empty dataframe
df = pd.DataFrame(columns=columns_order)
# data = {'Property URL': [], 'Title': [],'Price': [], 'Address': [], 'Description': [],
#         'Property Size':[], 'Land Area':[], 'Bedrooms':[], 'Bathrooms':[], 'Garages':[], 'Property Type':[], 'Property Status':[], 'Floor':[], 'Furnished':[] }
# df = pd.DataFrame(data)

while True:
  print(for_sale_url)
  for_sale_url = getnextpage(soup)
  if not for_sale_url:
    break
  salepage = requests.get(for_sale_url)
  soup = BeautifulSoup(salepage.text, 'html.parser')
  # Find all div elements with the specified class

  divs = soup.find_all('div', class_='d-flex align-items-center h-100')
  count = 1
    # Display the found divs
  for div in divs:
    # Extract information from the property div
    title = div.find('h2', class_='item-title').text.strip()
    price = div.find('li', class_='item-price').text.strip()
    address = div.find('address', class_='item-address').text.strip()

    link = div.find('a', class_='btn-item')['href']

    # Print or use the extracted information as needed
    # print(f"Title: {title}")
    # print(f"Price: {price}")
    # print(f"Address: {address}")

    if link:

      # print(f"Link: {link}")

      # Make another HTTP request to the linked page
      linked_page_response = requests.get(link)
      linked_page_html = linked_page_response.content

      # Parse the HTML of the linked page
      linked_page_soup = BeautifulSoup(linked_page_html, 'html.parser')

      # Find the description on the linked page
      description = linked_page_soup.find('div', class_='block-content-wrap').text.strip()
      # print("Description:", description)
      property_data = {
            'Property URL': link,
            'Title': title,
            'Type' : "For Sale",
            'Price': price,
            'Address': address,
            'Description': description,
            'Property Size': None,
            'Land Area': None,
            'Bedrooms': None,
            'Bathrooms': None,
            'Garages': None,
            'Property Type': None,
            'Property Status': None,
            'Floor': None,
            'Furnished': None,
            }

      ulists = linked_page_soup.find('div', class_='detail-wrap').find('ul', class_='list-2-cols list-unstyled')
      # finding all <li> tags
      details = ulists.find_all('li')
      for detail in details:
        label_tag = detail.find('strong')
        value_tag = detail.find('span')


        # Check if both label and value tags are present
        if label_tag and value_tag:
            label = label_tag.text.strip().replace(':', '')
            value = value_tag.text.strip()
            if label.lower() == 'garage' or label.lower() == 'garages':
                label = 'Garages'
            elif label.lower() == 'bathroom' or label.lower() == 'bathrooms':
                label = 'Bathrooms'
            elif label.lower() == 'bedroom' or label.lower() == 'bedrooms':
                label = 'Bedrooms'

            # print(label, value)
            if label in columns_order and label != 'Price':
              property_data[label] = value
              # print(property_data)
      # Append the data to the DataFrame
    df = pd.concat([df, pd.DataFrame([property_data])], ignore_index=True)






```
